Standardization of a dataset is a common requirement for many machine learning estimators: they might behave badly if the individual features do not more or less look like standard normally distributed data (e.g. Gaussian with 0 mean and unit variance).  
In this Exercise we will write a very simplified and naive feature normalizer

Given the function below called `normalizer` do the following step:
 - **ducumentation** (4pnts)
 - **function implementation** (12pnts)  
      Normalizing the elements of a 2D array:
    
            For each element of a column:
            Subtract the mean and devide it by variance (this is called Centering and scaling)

            Considering "i"th element of "j"th column of array X as x[i,j],
            transforme x[i,j] to z[i,j]:

            z[i,j] = (x[i,j] - u) / s

            where u is the mean of the column j and s is the standard deviation of that column.
 - **test your function** (4pnts). Do not forget edge cases
      
You can get some hints from folowing link:   
**But do not use it directly and write your code just with low level numpy and python commands**  
[sklearn.preprocessing.StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

```python
import numpy as np
def normalizer(X,):
    """
    Documentation:
        function description: the function subtracts the mean of the column from each element and divides each element by the variance
        inputs: 2D-array X[i,j]
        outputs: mean u, standard deviation s and 2D-array X_transformed[i,j]
    """
    ### Start of your code ##

    # check the inputs
    if X.ndim!=2:
      raise ValueError("wrong dimension of matrix X")
    # main body
    u=np.mean(X, axis=1)
    s=np.std(X, axis=1)
    #column or row wise, otherwise: axis=0??
    s[s==0]=1

    X_transformed=(X-u)/s
    ### End of your code ##
    return X_transformed


```

Write a function to transform features by scaling each feature to a given range.

```python
"""Documentation
function description: function scales each feature to a given range (e.g. [0,1]) columnwise
inputs: 2D array; given range e.g.(0,1)
outputs: scaled 2D array
"""
def scaler (X, given_range=(0,1)):
  # check the inputs
  if X.ndim!=2:
    raise ValueError("wrong dimension of matrix X")

  X_min=X.min(axis=0)
  X_max=X.max(axis=0)
  new_min,new_max=given_range

  #division by 0
  zero_range=X_min-X_max==0
  X_min[zero_range]=X_max[zero_range]-1

  #scaling
  X_scaled=((X-X_min)/(X_max-X_min))*(new_max-new_min)+new_min
  return X_scaled

#test01:
scaler(new_array)

#test02: array with minimum and maximum smaller than the given range
another_array=np.array([[0.3,0.55,0.6],[0.44,0.88,0.50],[0.2,0.3,0.4]])
print(another_array)
scaler(another_array)

#test03: array with wrong dimension
another_array2=np.arange(8).reshape(2,2,2)
print(another_array2)
scaler(another_array2)

#test04: array with one single value:
another_array3=np.ones(9).reshape(3,3)
print(another_array3)
scaler(another_array3)
```
